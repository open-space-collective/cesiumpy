```python
import os
import geopandas as gpd
```

```python
df = gpd.read_file(os.path.join('data', 'ca.json'))
df.head()
```

```python
g = df.loc[0, "geometry"]
g
```

```python
type(g)
```

```python
import cesiumpy
```

```python
v = cesiumpy.Viewer()
v.entities.add(cesiumpy.Wall(positions=g,
                    maximumHeights=10e5, minimumHeights=0,
                    material=cesiumpy.color.RED))
v
```
